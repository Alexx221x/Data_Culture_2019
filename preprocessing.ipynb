Читаем файлы

```
df = pd.read_excel('Выбросы в атмоферу стационарными и передвижными источниками.xls')
df1 = pd.read_excel("Затраты с 2000 по 2018.xls")
df2 = pd.read_excel("Количество осадков в 2018 году.xls")
df3 = pd.read_excel("Опасные явления.xls")
df4 = pd.read_csv("Природные черезвычайные ситуации.csv")
df5 = pd.read_excel("Совокупные выбросы парниковых газов.xlsx")
df6 = pd.read_excel("Температура по регионам.xls")
df7 = pd.read_excel('Выбросы_C02.xlsx')
df8 = pd.read_excel('Отходы_2010_2018.xls')
df9 = pd.read_excel('oxr_zatr1(2).xls')
```

```
display(df1.head())
display(df2.head())
display(df3.head())
display(df4.head())
display(df5.head())
display(df6.head())
display(df.head())
display(df7.head())
display(df8.head())
```

Обработка

```
df3 = df3.rename({'Unnamed: 0':'Год'},axis=1)
df = df.rename({'Unnamed: 1':'Год'},axis=1)
df8 = df8.rename({'Unnamed: 1':'Год'},axis=1)
df5 = df5.rename({'Unnamed: 0':'Год'},axis=1)
```

```
df9.head(2)
```

```
column = df9['Обновлено 30.08.2019']

col = df9[2:].head(1)
z = df9[4:].drop([5],axis=0).T
z.columns = z.head(1).values[0]
z = z[1:]
data = z.reset_index().drop('index', axis=1)
col = col.drop('Обновлено 30.08.2019', axis=1).values[0]
data['Год'] = col
data
```

Создание первого датафрейма 

```
df_main = df.merge(data, on='Год',how='outer').merge(df3,on='Год',how='outer').merge(df4,on='Год',how='outer').merge(df5,on='Год',how='outer').merge(df7,on='Год',how='outer').merge(df8,on='Год',how='outer')[10:]
df_main = df_main.drop(['Unnamed: 0_x','Unnamed: 0.1','Unnamed: 0_y'],axis=1) 
df_main
```

```
df_main.to_csv('Main_1.csv')
```
